# DATA PREPROCESSING

PLAN OF ACTION:
1. REMOVAL OF NULL VALUES
2. FEATURE ENCODING 
3. OUTLIER REMOVAL 
4. FIXING CLASS IMBALANCE
5. DROPPING IRRELEVANT COLUMNS

```python
# 1. Removing null values
for col in col_null_value:

    # Calculating mode
    mode = df[col].mode()[0]

    # Imputing null values with mode
    df[col] = df[col].fillna(mode)
```

```python
# 2. Feature Encoding
from sklearn.preprocessing import LabelEncoder 

encoder = LabelEncoder()

for col in obj_cols:
    df[col] = encoder.fit_transform(df[col])
```

```python
# 3.  Outlier Removal 

columns = df.columns.to_list()

for col in columns:
    
    if col in Normally_Distributed:
        mean = df[col].mean()
        std = df[col].std()
        upper_limit = mean + 3*std 
        lower_limit = mean - 3*std 
        df[col] = np.where(df[col]>upper_limit,upper_limit,np.where(df[col]<lower_limit,lower_limit,df[col]))
    
    elif col in Skewly_Distributed:
        percentile_25 = df[col].quantile(0.25)
        percentile_75 = df[col].quantile(0.75)
        IQR = percentile_75 - percentile_25
        upper_limit = percentile_75 + 1.5*IQR 
        lower_limit = percentile_25 + 1.5*IQR 
        df[col] = np.where(df[col]>upper_limit,upper_limit,np.where(df[col]<lower_limit,lower_limit,df[col])) 
```

```python
# 4. Dropping Irrelevant Column
df = df.drop(columns = ['customer_id','firstname','lastname'])
```

# FEATURE SELECTION

```python
# 1. Feature Set Using Mutual Information Score
from sklearn.feature_selection import mutual_info_classif 
from sklearn.feature_selection import SelectKBest 

y_final_close = df1.final_close_flag
y_primary_close = df1.primary_close_flag 
x = df1.drop(columns = ['final_close_flag','primary_close_flag'])

selector = SelectKBest(mutual_info_classif, k=45)

x_new = selector.fit_transform(x,y_final_close)
```

```python
selector.get_feature_names_out()
```

```python
feature_set_1 = pd.DataFrame(x_new, columns=selector.get_feature_names_out())
```

```python
feature_set_1
```

```python
feature_set_1.to_csv('Features_Credit_Risk_Analysis.csv')
```

# Data Splitting

```python
# Splitting dataset with Train Size (97.5% of total) & Test Size (2.5% of total)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(feature_set_1, y_final_close, test_size = 0.30,random_state=42)
```

```python
X_train.shape
```

# DATA TRANSFORMATION

```python
# Z-score based normalization
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.decomposition import PCA  

# captures 55% variance i.e. if there will be high variance there will be overfitting, so there has to be trade off b/w bias and variance 
pca = PCA(0.55) 

X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.fit_transform(X_test_scaled)
```

```python
X_train.shape
```

```python
X_train_pca.shape
```

# MODELLING

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier 
import tensorflow as tf 
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, LeakyReLU
from sklearn.metrics import classification_report, accuracy_score
```

```python
models = [LogisticRegression(max_iter=2000), DecisionTreeClassifier(), RandomForestClassifier()]
```

```python
# Without PCA with 45 features
for model in models:
    model.fit(X_train_scaled,y_train)
    print(f'Accuracy of {model} on train data is {100*model.score(X_train_scaled,y_train)}')
    print(f'Accuracy of {model} on test data is {100*model.score(X_test_scaled,y_test)}')
    y_pred = model.predict(X_test_scaled)
    print(classification_report(y_test,y_pred))
    print('*****************************************************************')
```

```python
# After PCA with 26 components
for model in models:
    model.fit(X_train_pca,y_train)
    print(f'Accuracy of {model} on train data is {100*model.score(X_train_pca,y_train)}')
    print(f'Accuracy of {model} on test data is {100*model.score(X_test_pca,y_test)}')
    y_pred = model.predict(X_test_pca)
    print(classification_report(y_test,y_pred))
    print('*****************************************************************')
```

```python
model = Sequential()

model.add(Dense(50,activation='ReLU',input_dim=45))
model.add(Dense(45, activation='ReLU'))
model.add(Dense(40, activation='ReLU'))
model.add(Dense(35, activation='ReLU'))
model.add(Dense(30, activation='ReLU'))
model.add(Dense(25, activation='ReLU'))
model.add(Dense(20, activation='ReLU'))
model.add(Dense(15, activation='ReLU'))
model.add(Dense(10, activation='ReLU'))
model.add(Dense(5, activation='ReLU'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Early Stopping is appliead to prevent unnecessary epochs,epoch after which validation loss begins to increase there early stopping will be called
callbacks = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=0,
    verbose=1,
    mode='min',
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0
)
```

```python
history = model.fit(X_train_scaled,y_train,validation_split=0.40,epochs=30,callbacks=callbacks)
```

```python
plt.plot(history.history['loss'],color='orange')
plt.plot(history.history['val_loss'],color='blue') 
plt.xlabel('Number Of Epochs')
plt.ylabel('Loss')
```

```python
plt.plot(history.history['accuracy'],color='orange')
plt.plot(history.history['val_accuracy'],color='blue') 
plt.xlabel('Number Of Epochs')
plt.ylabel('Accuracy')
```

Out of all models implemented, ANN gave best result i.e. there is no overfitting
